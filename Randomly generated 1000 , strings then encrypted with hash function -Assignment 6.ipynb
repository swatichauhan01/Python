# 3) Try to encrypt the randomly gererated- single string using hash function 
         

```python
strings=""
for i in range(string_length[0]):
    strings=strings+alphabets_array[np.random.randint(low=0,high=52,size=1)[0]]
    
print(strings)
    
num = int(input("Enter a number: "))
if num > 1:
    for j in range(2,num):
        if (num % j) == 0:
            break    #print(num,"is not a prime number and cannt encrypt")break  
        else:
            b=0
            for k in range(len(strings)):   #for k in range(len(str)):
                b=b + (ord(strings[k]))*(num**(k+1))
        
print(b,":" "Encypted Hash value of a string")  
#else:  
 #   print(num,"is not a prime number,we cannt encrypt msg")
    
```

# ii) Encryting 1000 strings with hash function 

```python
hash_values=[]
num = int(input("Enter a number: "))
if num > 1:
    for j in range(2,num):
        if (num % j) == 0:
            break    #print(num,"is not a prime number and cannt encrypt")break  
        else:
            b=0
            for i in range(len(strings)):
                for j in range(len(strings[i])):
                    b=b + (ord(strings[i][j]))*(num**(j+1))
                print(b,":""Encypted Hash value of a string") 
                hash_values.append(b)
                b=0
```
